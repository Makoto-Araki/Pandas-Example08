```python
# ライブラリのインポート
import pandas as pd
import japanize_matplotlib # pip install japanize_matplotlib

# CSVファイル読込、日付型項目を索引、時系列データ集計(parse_dates)
frame1 = pd.read_csv('C:\\Users\\Owner\\Pandas-Exercises8\FILE1.csv', index_col=['取引日'], parse_dates=True)
```

```python
# 時系列データを表示
frame1
```

```python
# 時系列データをグラフ化
frame1.plot()
```

```python
# グラフにタイトル付与
frame1.plot(title='取引量の推移')
```

```python
# グラフの色を緑・青・赤、点を四角、線を点線に変更
frame1.plot(title='取引量の推移', style=['gs-', 'bs-', 'rs-'])
```

```python
# 棒グラフに変更
frame1.plot(kind='bar', title='取引量の推移', rot=65)
```

```python
# 積上げ棒グラフに変更
frame1.plot(kind='bar', title='取引量の推移', rot=65, stacked=True)
```
